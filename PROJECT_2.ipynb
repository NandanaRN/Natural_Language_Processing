## NLP Project 2: Restaurant Feedback Analysis


### Problem Statement
A restaurant chain receives customer feedback through various channels, including food delivery apps, online review platforms, and feedback forms. Customers frequently share opinions on food taste, delivery speed, packaging, pricing, and customer service. Due to the increasing volume of feedback, restaurant managers struggle to manually analyze customer opinions and identify areas for improvement. The organization aims to leverage NLP techniques to analyze customer feedback, assess satisfaction levels, and pinpoint common issues affecting the dining experience.


### Tasks
1. Collect restaurant review data.
2. Preprocess customer feedback text.
3. Perform sentiment classification.
4. Identify common service issues.
5. Prepare a feedback analysis report.

### Create Sample Dataset

```python
data = {
    "Review": [
        "The food was delicious and fresh.",
        "Delivery was very late.",
        "Excellent customer service.",
        "Packaging was damaged.",
        "Food was okay.",
        "Loved the pizza and burger.",
        "Very expensive for the quality.",
        "Quick delivery and tasty food.",
        "The soup was cold.",
        "Amazing experience.",
        "Bad customer support.",
        "Average taste.",
        "Fresh ingredients and good quality.",
        "Order arrived after one hour.",
        "Nice ambience and friendly staff."
    ]
}

df = pd.DataFrame(data)

df
```

### Text Preprocessing

```python
stop_words = set(stopwords.words('english'))

def preprocess(text):

    text = text.lower()

    text = text.translate(str.maketrans('', '', string.punctuation))

    words = word_tokenize(text)

    words = [word for word in words if word not in stop_words]

    return " ".join(words)

df["Clean_Review"] = df["Review"].apply(preprocess)

df.head()
```

### Sentiment Analysis

```python
sia = SentimentIntensityAnalyzer()

def predict_sentiment(text):

    score = sia.polarity_scores(text)["compound"]

    if score >= 0.05:
        return "Positive"

    elif score <= -0.05:
        return "Negative"

    else:
        return "Neutral"

df["Sentiment"] = df["Clean_Review"].apply(predict_sentiment)

df
```

### Sentiment Count

```python
df["Sentiment"].value_counts()
```

### visualization: pie chart

```python
counts = df["Sentiment"].value_counts()

plt.figure(figsize=(6,6))

counts.plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Restaurant Feedback Sentiment")

plt.ylabel("")

plt.show()
```

### Find Common Issues

```python
negative_reviews = " ".join(
    df[df["Sentiment"]=="Negative"]["Clean_Review"]
)

words = negative_reviews.split()

common = Counter(words)

common.most_common(10)
```

### Word Cloud

```python
cloud = WordCloud(
    width=900,
    height=400,
    background_color="white"
).generate(negative_reviews)

plt.figure(figsize=(12,6))

plt.imshow(cloud)

plt.axis("off")

plt.title("Common Negative Issues")

plt.show()
```
